```python
patient_data.info()
```

Пропусков нет, ошибки в типе данных - есть 'active_before', 'edds_on' по смыслу числовые, но в датафрейме строковые ('object'). Необходимо приведение типов

```python
patient_data['active_before'] = (patient_data['active_before'].str.split()).apply(lambda x: float(x[0].replace(',', '.')))
patient_data['edds_on'] = (patient_data['edds_on'].str.split()).apply(lambda x: float(x[0].replace(',', '.')))
```

Распределения числовых столбцов

```python
for i in patient_data[['gps+','active_before','edds_on','anxiety','depression']].columns:
    f, (ax_hist, ax_box) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15,.85)})
    sns.boxplot(patient_data[i], ax=ax_hist, color="b")
    sns.kdeplot(patient_data[i], ax=ax_box,shade=True, color="b")
    # Remove x axis name for the boxplot
    ax_box.set(xlabel='')
    plt.show()
```

Тут можно сделать вывод по тому в каких пределах колеблетются значения депрессии и тревоги. Написать, что есть выбросы . Но пока их не буду удалять, потому что потеряю "крайне" депрессивных и "крайне" позитивных. Рассмотрим отдельно.

```python
#Категориальные признаки
for i in patient_data[['gps', 'sex', 'preparation', 'preparation_before',
       'obostr_do', 'injector', 'time', 'i_dh_p','laryngitis_tracheitis', 'i_moch_p', 'pyelrnfrith', 'herpes', 'aphthae',
       'weight_change', 'tonsillitis', 'pyelonephritis']].columns:
    ax = sns.countplot(x=patient_data[i], data=patient_data)
    plt.show()
```

Тут можно сделать общие какие-то выводы по группе. По тому сколько из группы с конкретным заболеванием, потере/набору веса, количеству мужчин/женщин в выборке, и.т.д.
